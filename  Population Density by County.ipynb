```python
import pandas as pd
import censusdata
from tabulate import tabulate
```

```python
df = censusdata.download('acs5', 
           2020,
           censusdata.censusgeo([('county', '*')]),
           ['B01003_001E'])
```

```python
df
```

```python
column_names = ['population']
df.columns = column_names
```

```python
df['percent_population'] = df.apply(
   lambda row: row['population']/332403.650, 
   axis = 1)
```

```python
df
```

```python
new_indices = []
county_names = []
for index in df.index.tolist():
        new_index = index.geo[0][1] + index.geo[1][1]
        new_indices.append(new_index)
        county_name = index.name.split(',')[0]
        county_names.append(county_name)
df.index = new_indices
df['county_name'] = county_names
```

```python
import plotly.express as px
from plotly.offline import plot

fig = px.choropleth(df,
                    locations='county_name', 
                    locationmode="USA-states", 
                    scope="usa",
                    color='percent_population',
                    color_continuous_scale="Viridis_r"
                    )
fig.layout.template = None
fig.show()
```

```python
fig = px.choropleth(df, geojson=county_names, locations='county_name', color='percent_population',
                           color_continuous_scale="Viridis",
                           range_color=(0, 12),
                           scope="usa",
                           labels={'pop':'population rate'}
                          )

fig.show()
```
